```python
# loop statement

# 1. for loop print numbers
for i in range(6):
    print(i)
numbers = [10, 20, 30, 40]
for num in numbers:
    print("list values:",num)

# 2. while loop
count = 1
while count <= 5:
    print("while loop:",count)
    count += 1

# 3. do-while loop
num = 1
while True:
    print("do-while loop:",num)
    num += 1
    if num > 5:
        break
```

```python
# Data Analytics Example
sales = [1000, 2500, 1800]
total = 0
for s in sales:
    total += s
print("Total Sales:", total)
```

```python
%%writefile calc1.py

from math import factorial
def add(a, b):
    return a + b
def sub(a, b):
    return a - b
def mul(a, b):
    return a * b
def div(a, b):
    return a / b
def pow(a, b):
    return a ** b
def fact(n):
    return factorial(n)
```

```python
from calc1 import add, sub, mul, div, pow, fact

print("Add 2 value:",add(5, 5))
print("sub 2 value:",sub(10, 5))
print("mul 2 value:",mul(5, 4))
print("div 2 value:",div(4, 8))
print("pow 2 value:",pow(2, 4))
print("fact value:",fact(5))
```
