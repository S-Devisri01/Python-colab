```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Dataset (slightly bigger)
df = pd.DataFrame({
    "hours": [1,2,3,4,5,6,7,8],
    "score": [35,40,50,60,70,75,85,90]
})

X = df[["hours"]]
y = df["score"]

# Train-test split (enough test data now)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

# Prediction (NO WARNING)
print("Predicted Score:",
      model.predict(pd.DataFrame({"hours": [9]}))[0])

# R2 Score (NO NaN)
y_pred = model.predict(X_test)
print("R2 Score:", r2_score(y_test, y_pred))

# Plot
plt.scatter(X, y)
plt.plot(X, model.predict(X), color="red")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Dataset
df = pd.DataFrame({
    "study_hours": [2, 4, 6, 8, 10, 12, 14, 16],
    "marks": [30, 35, 50, 60, 70, 75, 85, 95]
})

x = df[["study_hours"]]
y = df["marks"]

# Train-text-split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=42
)

model = LinearRegression()
model.fit(x_train, y_train)

# Prediction
print("Prediction Marks:",
      model.predict(pd.DataFrame({"study_hours":[18]}))[0])

# R2 Score
y_pred = model.predict(x_test)
print("R2 Marks:", r2_score(y_test, y_pred))

# Sort values for clean line
df_sorted = df.sort_values(by="study_hours")

plt.scatter(df["study_hours"], df["marks"], label="Actual")
plt.plot(
    df_sorted["study_hours"],
    model.predict(df_sorted[["study_hours"]]),
    color="red",
    label="Regression"
)

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Dataset (2 features)
df = pd.DataFrame({
    "study_hours": [2, 4, 6, 8, 10, 12, 14, 16],
    "sleep_hours": [8, 7, 7, 6, 6, 6, 5, 5],
    "marks": [35, 45, 55, 65, 72, 78, 88, 95]
})

# Features (Multiple)
X = df[["study_hours", "sleep_hours"]]
y = df["marks"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Model
model = LinearRegression()
model.fit(X_train, y_train)

# Prediction (NEW student)
new_data = pd.DataFrame({
    "study_hours": [10],
    "sleep_hours": [7]
})

print("Predicted Marks:", model.predict(new_data)[0])

# R2 Score
y_pred = model.predict(X_test)
print("R2 Score:", r2_score(y_test, y_pred))

# Plot
plt.scatter(df["study_hours"], df["marks"], label="Actual")

plt.plot(
    df["study_hours"],
    model.predict(df[["study_hours", "sleep_hours"]]),
    color="red",
    label="Regression (approx)"
)

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# ================= DATASET =================
df = pd.DataFrame({
    "size_sqft": [800, 1000, 1200, 1500, 1800, 2000, 2300, 2600],
    "bedrooms": [1, 2, 2, 3, 3, 4, 4, 5],
    "house_age": [20, 15, 10, 12, 8, 6, 5, 2],
    "price": [50, 65, 80, 95, 120, 140, 165, 190]  # in lakhs
})

# ================= FEATURES & TARGET =================
X = df[["size_sqft", "bedrooms", "house_age"]]
y = df["price"]

# ================= TRAIN TEST SPLIT =================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# ================= MODEL =================
model = LinearRegression()
model.fit(X_train, y_train)

# ================= PREDICTION =================
new_house = pd.DataFrame({
    "size_sqft": [1800],
    "bedrooms": [3],
    "house_age": [7]
})

print("Predicted House Price (Lakhs):", round(model.predict(new_house)[0], 2))

# ================= MODEL ACCURACY =================
y_pred = model.predict(X_test)
print("R2 Score:", round(r2_score(y_test, y_pred), 3))

# ================= VISUALIZATION (1 Feature) =================
plt.scatter(df["size_sqft"], df["price"], label="Actual Price")
plt.plot(
    df["size_sqft"],
    model.predict(df[["size_sqft", "bedrooms", "house_age"]]),
    color="red",
    label="Regression (Approx)"
)

plt.xlabel("House Size (sqft)")
plt.ylabel("Price (Lakhs)")
plt.title("House Price Prediction")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read File
df = pd.read_csv("/content/sales_data_assumption (1).csv")

# Feature & Target
X = df[["Advertising_Spend"]]
y = df["Sales"]

print(df.head())

# Train Model + Prediction
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)

df["Predicted_Sales"] = model.predict(X)
print("\nPrediction for 500 Ad Spend:",
      model.predict(pd.DataFrame({"Advertising_Spend":[500]}))[0])

# Residuals
df["Residuals"] = y - df["Predicted_Sales"]

# ASSUMPTION 1: LINEARITY
plt.scatter(X, y, label="Actual")
plt.plot(X, df["Predicted_Sales"], color="red", label="Regression Line")
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.legend()
plt.title("Linearity Check")
plt.show()

# ASSUMPTION 2: INDEPENDENCE (DurbinWatson)
from statsmodels.stats.stattools import durbin_watson
dw = durbin_watson(df["Residuals"])
print("Durbin-Watson:", dw)

# ASSUMPTION 3: HOMOSCEDASTICITY (Equal Spread)
plt.scatter(df["Predicted_Sales"], df["Residuals"])
plt.axhline(0, color="red")
plt.xlabel("Predicted Sales")
plt.ylabel("Residuals")
plt.title("Homoscedasticity Check")
plt.show()

# ASSUMPTION 4: NORMALITY (Q-Q Plot)
import scipy.stats as stats
stats.probplot(df["Residuals"], dist="norm", plot=plt)
plt.title("Q-Q Plot")
plt.show()

# KNN - Model Training + Prediction
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_scaled, y)
df["KNN_Predicted"] = knn.predict(X_scaled)
print("KNN Prediction:",
      knn.predict(scaler.transform([[500]]))[0])

# KNN Visualization
plt.scatter(X, y, label="Actual")
plt.scatter(X, df["KNN_Predicted"], color="green", label="KNN Prediction")
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.legend()
plt.title("KNN Prediction (Neighbor-based)")
plt.show()

# DECISION TREE - Train + Predict
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(max_depth=3)
dt.fit(X, y)
df["DT_Predicted"] = dt.predict(X)
print("Decision Tree Prediction:",
      dt.predict([[500]])[0])

# Decision Tree Visualization
plt.scatter(X, y, label="Actual")
plt.plot(X, df["DT_Predicted"], color="orange", label="Decision Tree")
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.legend()
plt.title("Decision Tree (Step-wise Prediction)")
plt.show()

# K-MEANS - Clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)

# K-Means Visualization
plt.scatter(X, y, c=df["Cluster"], cmap="viridis")
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.title("Customer / Sales Clustering")
plt.show()

#
```
