```python
# ============================================================
# ğŸ“˜ LSTM â€” Sentiment Analysis (IMDb)
# Goal: Understand how LSTM remembers word sequences
# ============================================================

import tensorflow as tf
import numpy as np

from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing import sequence


# ============================================================
# 1ï¸âƒ£ PARAMETERS
# ============================================================

VOCAB_SIZE = 10000     # Number of words to keep
MAX_LEN = 200          # Max sentence length
EMBED_DIM = 64         # Word embedding size
EPOCHS = 5             # Small value for demo


# ============================================================
# 2ï¸âƒ£ LOAD DATA â€” IMDb Movie Reviews
# ============================================================

(X_train, y_train), (X_test, y_test) = imdb.load_data(
    num_words=VOCAB_SIZE
)

# Pad sequences to same length
X_train = sequence.pad_sequences(X_train, maxlen=MAX_LEN)
X_test = sequence.pad_sequences(X_test, maxlen=MAX_LEN)

print("Training shape:", X_train.shape)


# ============================================================
# 3ï¸âƒ£ BUILD LSTM MODEL
# ============================================================

model = Sequential([
    # Converts word index â†’ dense vector
    Embedding(
        input_dim=VOCAB_SIZE,
        output_dim=EMBED_DIM,
        input_length=MAX_LEN
    ),

    # LSTM memory layer
    LSTM(64, return_sequences=False),

    # Regularization
    Dropout(0.5),

    # Output layer (binary classification)
    Dense(1, activation="sigmoid")
])

model.summary()


# ============================================================
# 4ï¸âƒ£ COMPILE MODEL
# ============================================================

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)


# ============================================================
# 5ï¸âƒ£ TRAIN MODEL
# ============================================================

history = model.fit(
    X_train,
    y_train,
    epochs=EPOCHS,
    batch_size=300,
    validation_split=0.2
)


# ============================================================
# 6ï¸âƒ£ EVALUATE MODEL
# ============================================================

loss, acc = model.evaluate(X_test, y_test)
print(f"\nâœ… Test Accuracy: {acc * 100:.2f}%")


# ============================================================
# 7ï¸âƒ£ TEST CUSTOM SENTENCES
# ============================================================

word_index = imdb.get_word_index()

def encode(text):
    tokens = [word_index.get(w, 2) for w in text.lower().split()]
    return sequence.pad_sequences([tokens], maxlen=MAX_LEN)


samples = [
    "The movie was fantastic",
    "The movie was boring and terrible",
    "Not a good movie",
    "good movie"
]

print("\nğŸ” Custom Predictions:")
for s in samples:
    pred = model.predict(encode(s), verbose=0)[0][0]
    sentiment = "Positive ğŸ˜€" if pred > 0.5 else "Negative ğŸ˜"
    print(f"{s} â†’ {sentiment} ({pred:.3f})")
```
