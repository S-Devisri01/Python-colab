```python
# ============================================================
# KERAS + TENSORFLOW WITH MANUAL STUDENT DATASET
# ============================================================
# Topics Covered:
# 1. Create Sequential Model
# 2. Compile Model
# 3. Model Callbacks (EarlyStopping, ModelCheckpoint)
# 4. Training and Evaluating Models
# 5. Saving and Loading Models
# 6. Custom Layers and Model Debugging
# 7. Monitoring with TensorBoard
# 8. Simple Prediction
# ============================================================

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import numpy as np

# ------------------------------------------------------------
# Step 1: Create Manual Student Dataset
# ------------------------------------------------------------
# Features: Marks in 4 subjects (out of 100)
# Target: 1 = Pass, 0 = Fail

np.random.seed(42)

X_train = np.array([
    [85, 90, 78, 92],   # Student 1
    [40, 50, 45, 60],   # Student 2
    [70, 75, 80, 65],   # Student 3
    [30, 25, 40, 35],   # Student 4
    [90, 88, 95, 92],   # Student 5
    [55, 60, 58, 62],   # Student 6
    [20, 35, 30, 25],   # Student 7
    [65, 70, 60, 75],   # Student 8
    [50, 55, 45, 60],   # Student 9
    [95, 100, 98, 97]   # Student 10
])

y_train = np.array([1, 0, 1, 0, 1, 1, 0, 1, 0, 1])

# ------------------------------------------------------------
# Step 2: Create Sequential Model
# ------------------------------------------------------------
model = Sequential([
    layers.Dense(8, activation='relu', input_shape=(4,)),
    layers.Dense(1, activation='sigmoid')
])

# ------------------------------------------------------------
# Step 3: Compile Model
# ------------------------------------------------------------
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ------------------------------------------------------------
# Step 4: Add Callbacks
# ------------------------------------------------------------
early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=2,
    restore_best_weights=True
)

save_best = keras.callbacks.ModelCheckpoint(
    'best_student_model.keras',
    save_best_only=True
)

tensorboard_cb = keras.callbacks.TensorBoard(
    log_dir='./logs',
    histogram_freq=1
)

# ------------------------------------------------------------
# Step 5: Train Model
# ------------------------------------------------------------
history = model.fit(
    X_train,
    y_train,
    epochs=20,
    validation_split=0.2,
    callbacks=[early_stop, save_best, tensorboard_cb],
    verbose=1
)

# ------------------------------------------------------------
# Step 6: Evaluate Model
# ------------------------------------------------------------
loss, acc = model.evaluate(X_train, y_train, verbose=0)
print(f"âœ… Final Accuracy on Student Dataset: {acc:.2f}")

# ------------------------------------------------------------
# Step 7: Save & Load Model
# ------------------------------------------------------------
model.save("student_model.keras")
loaded_model = keras.models.load_model("student_model.keras")
print("âœ… Student model saved and loaded successfully!")

# ------------------------------------------------------------
# Step 8: Custom Layer Example
# ------------------------------------------------------------
class MultiplyByTwo(layers.Layer):
    def call(self, inputs):
        return inputs * 2

custom_model = Sequential([
    MultiplyByTwo(),
    layers.Dense(4, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

custom_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

custom_model.fit(X_train, y_train, epochs=5, verbose=1)

loss, acc = custom_model.evaluate(X_train, y_train, verbose=0)
print(f"âœ… Custom layer model accuracy: {acc:.2f}")

# ------------------------------------------------------------
# Step 9: Simple Prediction
# ------------------------------------------------------------
new_student = np.array([[75, 80, 70, 85]])

prediction = model.predict(new_student)
print("ğŸ¯ Predicted Pass Probability:", prediction[0][0])
```
