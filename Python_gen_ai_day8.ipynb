```python
# Data Visualization using Seaborn

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the "tips" dataset
tips = sns.load_dataset("tips")

# Show first 5 rows
print(tips.head())

# =========================
# SCATTER PLOT
# =========================
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()

# =========================
# HISTOGRAM
# =========================
sns.histplot(tips["total_bill"], bins=30, kde=True)
plt.title("Histogram of Total Bill")
plt.show()

# =========================
# BOX PLOT
# =========================
sns.boxplot(x="smoker", y="tip", data=tips)
plt.title("Box Plot: Tips by Smoker")
plt.show()

# =========================
# BAR PLOT
# =========================
sns.barplot(x="day", y="tip", data=tips)
plt.title("Bar Plot: Average Tips per Day")
plt.show()

# =========================
# CORRELATION HEATMAP
# =========================
# Select only numerical columns
numerical_tips = tips.select_dtypes(include=["float64", "int64"])
corr = numerical_tips.corr()

# Heatmap - Coolwarm
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Coolwarm)")
plt.show()

# Heatmap - Viridis
sns.heatmap(corr, annot=True, cmap="viridis")
plt.title("Correlation Heatmap (Viridis)")
plt.show()

# Heatmap - Blues
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title("Correlation Heatmap (Blues)")
plt.show()

# Heatmap - Greens
sns.heatmap(corr, annot=True, cmap="Greens")
plt.title("Correlation Heatmap (Greens)")
plt.show()

# =========================
# PAIRPLOT
# =========================
sns.pairplot(tips, hue="smoker")
plt.show()

# =========================
# CUSTOMIZED SCATTER PLOT
# =========================
sns.set_style("darkgrid")
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker",
    data=tips
)
plt.title("Scatter Plot with Style")
plt.show()
```
