```python
import pandas as pd

df = pd.read_csv("/content/movies.csv")
print(df.head())
print(df.info())
print(df.isnull().sum())
df = df.drop_duplicates()

```

```python
# Numeric columns → use mean
df["score"].fillna(df["score"].mean(), inplace=True)
df["votes"].fillna(df["votes"].mean(), inplace=True)
df["budget"].fillna(df["budget"].mean(), inplace=True)
df["gross"].fillna(df["gross"].mean(), inplace=True)
df["runtime"].fillna(df["runtime"].mean(), inplace=True)
# Categorical / non-numeric columns → use mode
df["rating"].fillna(df["rating"].mode()[0], inplace=True)
df["released"].fillna(df["released"].mode()[0], inplace=True)
df["writer"].fillna(df["writer"].mode()[0], inplace=True)
df["star"].fillna(df["star"].mode()[0], inplace=True)
df["country"].fillna(df["country"].mode()[0], inplace=True)
df["company"].fillna(df["company"].mode()[0], inplace=True)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop("budget", axis=1) #input
y = df["budget"] #output

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=45)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

```
